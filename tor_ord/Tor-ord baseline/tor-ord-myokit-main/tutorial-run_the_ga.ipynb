## Run a Tor-ORd parameter recovery genetic algorithm

---

**Overview:** In this tutorial, you will run a genetic algorithm that recovers the parameters of the Tor-ORd model. The baseline Tor-ORd model has a value of 1.0 for all of the following conductances: GCaL, GKs, GKr, GNaL, and Jup. The goal of the GA is to fit the Tor-ORd model to a single AP from the baseline version of the Tor-ORd model, and hopefully recover the parameters.

All of the code used in this simulation can be found in `supercell_ga`.

```python
import supercell_ga
```

**The start_ga function:** The following code is used to run the GA. This function can accept a population size (e.g. `pop_size=10`) and number of generations (e.g. `max_generations=3`) as arguments. After each generation of the GA is run, it will print out the average fitness of all individuals in that generation, and the fitness of the best individual.

**So, what is fitness?**

In this GA, we define fitness as the difference in AP (`dvdt_max`, `apd_10`, `apd_50`, `apd_90`) and CaT (`cat_amp`, `cat_10`, `cat_50`, `cat_90`) features between an individual and the baseline Tor-ORd model. A perfect fitness score is equal to zero, and represents when an individual's AP and CaT features perfectly match with the original Tor-ORd features. 

When you're finished running the simulation, `supercell_ga.plot_generation()` will plot the parameter values for the 10 best individuals, and APs for the best and the original models. Because the population size and number of generations are so small, your result will likely be far from a perfect fit.

```python
all_individuals = supercell_ga.start_ga(pop_size=10, max_generations=3)

supercell_ga.plot_generation(all_individuals, gen=None, is_top_ten=True)
```

**Increase generations and population:** If you are up for waiting a few minutes, the following code will run a simulation with 50 individuals per generation and five generations. You will likely see the best fitness values drop, but the results will still be short of perfect.

```python
all_individuals = supercell_ga.start_ga(pop_size=50, max_generations=5)

supercell_ga.plot_generation(all_individuals, gen=None, is_top_ten=True)
```
